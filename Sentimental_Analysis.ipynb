```python
import numpy as np
```

```python
import pandas as pd
```

```python
path='/content/Reddit_Data.csv'
```

```python
df=pd.read_csv(path)
```

```python
df.shape
```

```python
df.head()
```

```python
df.category.value_counts()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.category.value_counts()
```

```python
min_size=8277
df1=df[df['category']==1].sample(min_size,random_state=2025)
df2=df[df['category']==0].sample(min_size,random_state=2025)
df3=df[df['category']==-1].sample(min_size,random_state=2025)
```

```python
df_balanced=pd.concat([df1,df2,df3],axis=0)
```

```python
df_balanced.head()
```

```python
df_balanced.tail()
```

```python
df_balanced.category.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(df_balanced['clean_comment'],df_balanced['category'],stratify=df_balanced['category'],random_state=2025)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=Pipeline([('tfidf',TfidfVectorizer()),('r',RandomForestClassifier())])
```

```python
clf.fit(X_train,Y_train)
```

```python
Y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
clf2=Pipeline([('tfidf',TfidfVectorizer()),('lr',LogisticRegression(multi_class='multinomial', class_weight='balanced'))])
```

```python
clf2.fit(X_train,Y_train)
```

```python
Y_pred=clf2.predict(X_test)
```

```python
print(classification_report(Y_test,Y_pred))
```
